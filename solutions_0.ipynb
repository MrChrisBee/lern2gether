## 6. Quader: Schreibe ein Programm, bei dem der Nutzer Breite,
### Länge und Höhe eines Quaders angibt. Das Programm
### berechnet dann das Volumen, die Oberfläche und die
### Raumdiagonale ( √ a2 +b2+ c2 )

```python
import math
breite = float(input("Bitte die Breite eingeben"))
laenge = float(input("Bitte die Länge eingeben"))
hoehe = float(input("Bitte die Höhe eingeben"))
volumen = laenge * breite * hoehe
oberflaeche = 4 * (hoehe * breite) + 2 * (laenge * breite)
raumdiagonale = math.sqrt(laenge **2 + breite **2 + hoehe **2)
print(f"Das Volumen ist {volumen} Einheiten")
print(f"Die Oberfläche ist {oberflaeche} Einheiten")
print(f"Die Raumdiagonale ist {raumdiagonale} Einheiten")
```

## 7. Quaderberechnung abändern: Ändere das Programm

### falls er negative Eingaben macht, dass dies nicht sein kann.

```python
import math
breite = float(input("Bitte die Breite eingeben"))
laenge = float(input("Bitte die Länge eingeben"))
hoehe = float(input("Bitte die Höhe eingeben"))
if (breite < 0 or laenge < 0 or hoehe < 0 ):
    print("Bitte nur positive Werte eingeben!")
else:
    volumen = laenge * breite * hoehe
    oberflaeche = 4 * (hoehe * breite) + 2 * (laenge * breite)
    raumdiagonale = math.sqrt(laenge **2 + breite **2 + hoehe **2)
    print(f"Das Volumen ist {volumen} Einheiten")
    print(f"Die Oberfläche ist {oberflaeche} Einheiten")
    print(f"Die Raumdiagonale ist {raumdiagonale} Einheiten")
```

## 9. Rückgeld: 
### Schreibe ein Programm, bei dem man eingibt, wie
### viel jemand zahlen muss und wie viel er in bar gezahlt hat. Es
### soll ausgegeben werden, wie viel Rückgeld der Kunde
### bekommt. Falls er zu wenig bezahlt hat, soll das Programm
### eine Warnung ausgeben

```python
zuzahlen, bar = 0.0, 0.0
zuzahlen = float(input("Was musst der Kunde Zahlen? "))
bar = float(input("Was hat der Kunde bezahlt? "))
if (bar < zuzahlen):
    print(f"Da fehlen noch {zuzahlen-bar} Einheiten! ")
elif bar == zuzahlen:
    print("Nett lächeln und danke sagen!")
else:
    print(f"Zahle dem Kunden bitte {bar-zuzahlen} Einheiten zurück. ")
```

## 10. Das Programm soll eine quadratische Gleichung der Forma x 2+b x +c=0 lösen können. 
### Der Nutzer gibt hierfür die Zahlen a, b, c ein. 
### Das Programm prüft die Diskriminante, also b^2−4ac, ob diese positiv, null oder negativ ist und gibt dann
### aus wie viele Lösungen es gibt und ggf. deren Werte.

```python
print("Lösung der Gleichung ax²+bx+c=0 ")
a = float(input("Gib mir bitte einen Wert für a an"))
b = float(input("Gib mir bitte einen Wert für b an"))
c = float(input("Gib mir bitte einen Wert für c an"))
diskriminante = b**2-4*a*c
if diskriminante < 0: #kleiner keine
    print("Die Kurve schneidet die X Achse nicht. Keine Lösung")
elif diskriminante == 0:
    print("Die Kurve schneidet die X Achse an einer Stelle")
    n1 = (-b+(diskriminante**0.5))/(2*a)
    print(f"Die Nullstelle ist: {n1}")
else:
    n1 = (0-b+(diskriminante**0.5))/(2*a)
    print(f"Eine Nullstelle ist: {(0-b+(diskriminante**0.5))/2*a}")
    print(f"Eine Nullstelle ist: {(0-b-(diskriminante**0.5))/2*a}")
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Start: -100
# End: 100
# Steps: 0.1
x = np.arange(-100, 100, 0.1)
y = a * x**2 + b * x + c

plt.xlabel('x-Werte')
plt.ylabel('y-Werte')
plt.title(f'Funktion ${a}x^2 + {b}x + {c}$')

plt.plot(x, y)
plt.show()
```

## 19. Größte Zweierpotenz: 
### Schreibe ein Programm, bei dem der Nutzer eine natürliche Zahl eingibt und 
### das Programm gibt die größte Zweierpotenz zurück, 
### die kleiner oder gleich der eingegebenen Zahl ist.

```python
grenze = int(input("Natürliche Zahl"))
wert = 0
while (wert ** 2) <= grenze:
    wert += 1
print(f"Wert: {wert-1}² = {(wert-1)**2}")
```
